### Option 1: Use Pandas to create the contacts DataFrame.

```python
contact_info_df.columns = contact_info_df.iloc[0]
contact_info_df = contact_info_df[1:]

contact_info_df.head()
```

```python
type(contact_info_df['contact_info'].values.tolist()[0])
```

```python
import json
json.loads(contact_info_df['contact_info'].values.tolist()[0]).values()
```

```python
# Iterate through the contact_info_df and convert each row to a dictionary.
import json
dict_values = [list(json.loads(i).values()) for i in contact_info_df['contact_info'].values.tolist()]


# Print out the list of values for each row.
print(len(dict_values))
print(dict_values[0:100])
```

```python
# Iterate through the contact_info_df and convert each row to a dictionary.
# import json
# dict_values = []


# # Print out the list of values for each row.
# print(dict_values)
```

```python
# Create a contact_info DataFrame and add each list of values, i.e., each row 
# to the 'contact_id', 'name', 'email' columns.
contacts_df = pd.DataFrame()

contacts_df['contact_id'] = [contact[0] for contact in dict_values]
contacts_df['name'] = [contact[1] for contact in dict_values]
contacts_df['email'] = [contact[2] for contact in dict_values]

contacts_df.head()
```

```python
# Check the datatypes.
contacts_df.dtypes
```

```python
# Create a "first"name" and "last_name" column with the first and last names from the "name" column. 
contacts_df['first_name'] = [name.split(" ")[0] for name in contacts_df['name']]
contacts_df['last_name'] = [name.split(" ")[1] for name in contacts_df['name']]

# Drop the contact_name column
contacts_firstlast_df = contacts_df.drop(columns='name')

contacts_firstlast_df.head()
```

```python
# Reorder the columns
contacts_df_clean = contacts_firstlast_df[["contact_id", "first_name", "last_name", "email"]]

contacts_df_clean.head(10)
```

```python
# Check the datatypes one more time before exporting as CSV file.
contacts_df_clean.info()
```

```python
# Export the DataFrame as a CSV file. 
contacts_df_clean.to_csv("Resources/contacts.csv", encoding='utf8', index=False)
```

### Option 2: Use regex to create the contacts DataFrame.

```python
contact_info_df_copy = contact_info_df.copy()
contact_info_df_copy.head()
```

```python
# Extract the four-digit contact ID number, convert to integer, add to contact_info_df_copy dataframe
import re

contact_list = contact_info_df_copy.values.tolist()
contact_ids = [int(re.search('\d{4}', contact_list[i][0]).group(0)) for i in range(len(contact_list))]
contact_info_df_copy['contact_id'] = contact_ids

contact_info_df_copy.head()
```

```python
# Check the datatypes.
contact_info_df_copy.info()
```

```python
# Extract the name of the contact and add it to a new column.
contact_names = [re.search('[a-zA-Z]+ [a-zA-Z]+', contact_list[i][0]).group(0) for i in range(len(contact_list))]
contact_info_df_copy['name'] = contact_names

contact_info_df_copy.head(10)
```
